### 단계 구분도

```python
import pandas as pd
# 인구 데이터
df = pd.read_excel('/content/drive/MyDrive/lsda02/data/경기도인구데이터.xlsx',
                   index_col = '구분', engine = 'openpyxl')
df.head()
```

```python
import json
# 지도 데이터 = 경기도 경계선 폴리곤
bnd = json.load(open('/content/drive/MyDrive/lsda02/data/경기도행정구역경계.json',
                encoding = 'utf-8'))
```

```python
# 키 확인
bnd.keys()
```

```python
bnd['features']
```

```python
# 값 확인
bnd['features'][0]['properties']['name']
```

```python
df.columns
```

```python
# 열 이름 데이터 타입을 문자열로 변환
df.columns = df.columns.map(str)
```

```python
df.columns
```

```python
# 경기도 지도 만들기
ggd_map = folium.Map(location = [37.55, 126.98], zoom_start = 9)
# 단계 구분도 - 2007년
year = "2007"
folium.Choropleth(geo_data = bnd, # 지도 데이터 = 경기도 경계선
                  data = df[year], # 인구 데이터 = 색깔로 표현할 값
                  columns = [df.index, df[year]], # 지역의 이름, 해당 연도 인구
                  fill_color = 'Reds', # ColorBrewer code
                  fill_opacity = 0.3, # 다각형 내부 색상 투명도
                  line_opacity = 0.5, # 선 투명도
                  # 지도 데이터와 인구 데이터를 연결하는 내용 = 지역 이름
                  # 원래 features 이지만, 공식 문서 문법 표기상 feature 사용
                  key_on = 'feature.properties.name').add_to(ggd_map)
# 파일로 저장
ggd_map.save('./ggd_' + year + '.html')
```

### 인덱스

```python
# 이름 = 인덱스 이름 / 라벨
test_dict = {
    'a': 1,
    'b': 2,
    'c': 3
}
sr = pd.Series(test_dict)
# 인덱스 확인
print(sr.index)
# 데이터 타입이 object = 문자
# 인덱스 이름 = a, b, c
```

```python
# 값 확인
print(sr.values)
# 딕셔너리의 키 => 인덱스 이름
# 딕셔너리의 값 => 값 = 숫자 int
```

```python
# 숫자 = 정수형 위치 인덱스
# 리스트 생성
# 기본자료형 = 문자열, 숫자, 불
test_list = ["Hello", 23, 3.14, True]
print(test_list)
```

```python
# 리스트를 시리즈로 변환
sr = pd.Series(test_list)
print(sr)
```

```python
# 인덱스 확인
print(sr.index)
# 값 확인
print(sr.values)
# 리스트의 인덱스 => 인덱스 숫자(정수)
# 리스트의 요소 => 값 = 기본자료형
```

### 원소 선택

```python
# 인덱스 사용
# 튜플 생성
# 리스트와 유사
# 차이점 : 소괄호 사용, 원소(개별 값) 변경 불가
# 기본자료형
test_tuple = ("정대만", 184, 73.5, True)
print(test_tuple)
```

```python
# 튜플을 시리즈로 변환
sr = pd.Series(test_tuple, index = ['이름','키','체중','불꽃남자'])
print(sr)
```

```python
# 하나의 원소 선택
# 숫자
print(sr[0]) # 이름
print(sr[1]) # 키
print(sr[2]) # 체중
print(sr[3]) # 불꽃남자
```

```python
# 이름
print(sr['이름']) # 이름
print(sr['키']) # 키
print(sr['체중']) # 체중
print(sr['불꽃남자']) # 불꽃남자
```

```python
# 여러 개의 원소 선택 = 리스트 형태로 입력
# 숫자
print(sr[[0,3]]) # 이름, 불꽃남자
print(sr[[1,3]]) # 키, 불꽃남자
# 이름
print(sr[['이름', '체중']]) # 이름, 체중
print(sr[['키', '불꽃남자']]) # 키, 불꽃남자
```

```python
# 범위를 지정하여 여러 개의 원소 선택 = 슬라이싱
# 숫자 = 끝 번호 포함하지 않음
print(sr[0:2]) # 이름, 키
print(sr[0:1+1]) # 이름, 키
print(sr[1:4]) # 키, 체중, 불꽃남자
print(sr[1:3+1]) # 키, 체중, 불꽃남자
# 이름
print(sr['이름':'키']) # 이름, 키
print(sr['키':'불꽃남자']) # 키, 체중, 불꽃남자
```
